## 2. A3C-LSTM Agent design

```python
# import Gym and the Trading Environment
import gym
import gym_tradesim
gym.envs.registry.all()
```

```python
# Load TradeSim environment
env = gym.make('TradeSim-v0')
```

At this point the environment is loaded and dataframe is ready. Now let's modify environment callbacks for data ingestion (simulating data provided by the Trading Platform and the LSTM Predictive model):

```python
# modify callback for prediction data ingestion (get data from dataframe in stead from LSTM model)
def pull_predicted_data(step): 
  return {'high':df_a3c.iloc[step]['PREDICT_HIGH_PRICE'], 'low':df_a3c.iloc[step]['PREDICT_LOW_PRICE']}

# modify callback for real-time quotes data ingestion (get data from dataframe in stead from TP)
def pull_last_tick(instrument, step): 
  return {'ask':df_a3c.iloc[step]['ASK'], 'bid':df_a3c.iloc[step]['BID']}

# environment configuration:
# Max steps per episode = Num rows of df_a3c dataframe
# Max account balance = 10000€
# Max profit per episode = 10000€
# callback attachments
# callback for market info delegated to TradeSim-v0 environment
env.env.configure(steps_per_episode=df_a3c.shape[0],      
                  max_price=5 * np.max(df_a3c['ASK'].values), 
                  max_balance=10000.0, 
                  max_pl=10000.0, 
                  cb_pull_predictions=pull_predicted_data, 
                  cb_pull_ticks=pull_last_tick,
                  cb_market_info=None)
```

```python
# reset and check initial state
obs = env.reset()
obs
```

```python
# executes random actions until the end of the episode
obs = env.reset()
done = False

while not done:
  action=env.action_space.sample()
  state, reward, done, info = env.step(action)
  
print('Episode End!')

# get episodic stats
stats = env.env.statistics()

# Visualize them
plt.figure(figsize=(14,20))
plt.subplot(4,1,1)
plt.plot(np.asarray(stats['rewards']))
plt.legend(['rewards'])
plt.subplot(4,1,2)
plt.plot(np.cumsum(np.asarray(stats['rewards'])))
plt.legend(['total rewards'])
plt.subplot(4,1,3)
plt.plot(np.asarray(stats['balances']))
plt.legend(['balance'])

# Draw a price series as mid point between Bid,Ask
price = (df_a3c['BID'].values + df_a3c['ASK'].values)/2
price = price[1:]

# Get indexes for operation markers
x_marker_open_long = np.where(np.asarray(stats['pos_open_long']) > 0)[0]
x_marker_open_short = np.where(np.asarray(stats['pos_open_short']) > 0)[0]
x_marker_close = np.where(np.asarray(stats['pos_close']) > 0)[0]

# Get values for operation markers
y_marker_open_long = np.asarray(stats['pos_open_long']) * price[:np.asarray(stats['pos_open_long']).shape[0]]
y_marker_open_short = np.asarray(stats['pos_open_short']) * price[:np.asarray(stats['pos_open_short']).shape[0]]
y_marker_close = np.asarray(stats['pos_close']) * price[:np.asarray(stats['pos_close']).shape[0]]
y_marker_open_long = np.take(y_marker_open_long, x_marker_open_long)
y_marker_open_short = np.take(y_marker_open_short, x_marker_open_short)
y_marker_close = np.take(y_marker_close, x_marker_close)

plt.subplot(4,1,4)
# Show price
plt.plot(price, color='blue')
# Show opened long-operations markers
plt.scatter(x_marker_open_long, y_marker_open_long, marker='^', color='g', s=100)
# Show opened short-operations markers
plt.scatter(x_marker_open_short, y_marker_open_short, marker='v', color='r', s=100)
# Show close operations markers
plt.scatter(x_marker_close, y_marker_close, marker='o', color='black', s=30)
plt.show()
```

Now let's provide a cell to execute a whole episode using Agent actions. Modifying variable `NUM_EPISODES` can execute multiple episodes in the same cell execution.

```python
# reset the environment
obs = env.reset()
done = False
NUM_EPISODES = 1
_episodes = 0
results={}

while _episodes < NUM_EPISODES:
  while not done:
    # Agent selects an action for current observation
    action = agent.selectAction(obs)
    # Executes the action
    state, reward, done, info = env.step(action)
    # Run Agent's learning process
    agent.learn(action, state, reward, done)
    
  # get episodic stats
  stats = env.env.statistics()
  # save episodic results
  results.append(stats)
  
  _episodes += 1
  
print('Episode loop Ends!')
```
