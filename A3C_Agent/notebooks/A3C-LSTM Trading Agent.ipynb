#### **Data Merging**

Ok, at this point I'm ready to build ```df_a3c``` dataframe, merging all these preloaded data. As each dataset has different timeframe: 

- price predictions in 4-hours timeframe
- candlesticks patterns in 30-min timeframe 
- real-time quotes in 5-min timeframe

I have to add them accordingly, using the smallest as timeframe base (that is 5-min). For that purpose I'm going to execute lambda functions on price predictions and candlestick patterns to adapt to 5-min timeframe. Let's go:

````python
# First I must ensure that first and last rows in df_rt has the same timestamp as df_predict:

# get limits
first_valid_h4_timestamp = df_predict.iloc[0].name.strftime("%Y-%m-%d %H:%M:%S")
last_valid_h4_timestamp = (df_predict.iloc[-1].name + pd.DateOffset(hours=4)).strftime("%Y-%m-%d %H:%M:%S")
print('Set limits between: ', first_valid_h4_timestamp, ' and ', last_valid_h4_timestamp)

# get a slice of df_rt according with such limits
df_rt = df_rt[(df_rt.index >= first_valid_h4_timestamp) & (df_rt.index < last_valid_h4_timestamp)]
print('df_rt sliced from: ', df_rt.iloc[0].name, ' to ', df_rt.iloc[-1].name)

# Execute lambda functions for H4 price predictions conversion to M5 timeframe
df_rt['PREDICT_HIGH_PRICE'] = df_rt.apply(lambda row: df_predict[df_predict.index==row.name.replace(hour=row.name.hour-(row.name.hour%4), minute=0, second=0)]['PREDICT_HIGH_PRICE'].values[0] , axis=1)
df_rt['PREDICT_LOW_PRICE'] = df_rt.apply(lambda row: df_predict[df_predict.index==row.name.replace(hour=row.name.hour-(row.name.hour%4), minute=0, second=0)]['PREDICT_LOW_PRICE'].values[0] , axis=1)
df_rt.head()
````

````python
# Now proceed in the same way with candlestick pattterns

# get limits
first_valid_m30_timestamp = df_cdl.iloc[0].name.strftime("%Y-%m-%d %H:%M:%S")
last_valid_m30_timestamp = (df_cdl.iloc[-1].name + pd.DateOffset(minutes=30)).strftime("%Y-%m-%d %H:%M:%S")
print('Set limits between: ', first_valid_m30_timestamp, ' and ', last_valid_m30_timestamp)

# get a slice of df_rt according with such limits
df_rt = df_rt[(df_rt.index >= first_valid_m30_timestamp) & (df_rt.index < last_valid_m30_timestamp)]
print('df_rt sliced from: ', df_rt.iloc[0].name, ' to ', df_rt.iloc[-1].name)

# Execute lambda functions for M30 candlestick patterns conversion to M5 timeframe
for c in df_cdl.columns:
  df_rt[c] = df_rt.apply(lambda row: df_cdl[df_cdl.index==row.name.replace(minute=row.name.minute-(row.name.minute%30), second=0)][c].values[0] , axis=1)

df_rt.head(3)
````

There is only one more step. Select only quotes, candlesticks and predictions. Then once selected, we build a new dataframe named `df_a3c`.

````python
# import Gym and the Trading Environment. Check if environment is loaded at the end of the output log
import gym
import gym_tradesim
gym.envs.registry.all()
````

````python
# Load TradeSim environment
env = gym.make('TradeSim-v0')
````

````python
# Callback setup to get Price predictions from df_a3c, simulating the 1st-stage Predictive Model
def pull_predicted_data(step): 
  return {'high':df_a3c.iloc[step]['PREDICT_HIGH_PRICE'], 'low':df_a3c.iloc[step]['PREDICT_LOW_PRICE']}

# Callback setup to get real-time quotes from df_a3c, simulating the Broker's Trading Platform
def pull_last_tick(instrument, step): 
  return {'ask':df_a3c.iloc[step]['ASK'], 'bid':df_a3c.iloc[step]['BID']}

# Callback setup to get real-time quotes from df_a3c, simulating the Broker's Trading Platform
# previously is required a cdl_columns selection:
cdl_cols = []
for c in df_a3c.columns:
  if c.startswith('CDL'):
    cdl_cols.append(c)

def pull_last_cdl_patterns(step): 
  cdles = df_a3c[cdl_cols]
  return np.asarray(cdles.iloc[step])
````

````python
# Create dataframe slices for training and validation
train_val_ratio = 0.8
training_sample_count = int(df_a3c.shape[0] * train_val_ratio)

df_a3c_training   = df_a3c[0:training_sample_count]
df_a3c_validation = df_a3c[training_sample_count:]
````

````python

# Environment configuration by internal method 'configure'
# Episode's steps = df_a3c rows
# instrument = Financial instrument descriptive name
# equity = Initial account equity value
# leverage = Leverage used per operation
# commissions = Fixed commissions applied after each operation
# num_cdl_patterns = Number of candlestick pattern detectors
# cb_pull_cdl_patterns = Callback to get candlestick pattern detectors. 
# cb_pull_predictions = Callback to get predictions. 
# cb_pull_ticks = Callback to get quotes. Ej:
# cb_market_info = None (internal emulator will be used)        
env.env.configure(steps_per_episode=df_a3c_training.shape[0],      
                  instrument = 'EURUSD', 
                  equity=10000.0, 
                  leverage=1.0, 
                  commissions=0.5,
                  num_cdl_patterns=len(cdl_cols),
                  cb_pull_cdl_patterns=pull_last_cdl_patterns,
                  cb_pull_predictions=pull_predicted_data, 
                  cb_pull_ticks=pull_last_tick,
                  cb_market_info=None)
````

````python
####################################################################################
####################################################################################
# Test Execution
####################################################################################
####################################################################################

####################################################################################
# Global Constants:

# Execution time in seconds
RUN_TIME = 30
# Parallel threads
THREADS = 8
# Num optimizers
OPTIMIZERS = 2
# Delays between threads
THREAD_DELAY = 0.001

# A3C learning hyper-parameters
GAMMA = 0.99
N_STEP_RETURN = 8
GAMMA_N = GAMMA ** N_STEP_RETURN
EPS_START = 0.4
EPS_STOP  = .15
EPS_STEPS = df_a3c_training.shape[0]
MIN_BATCH = 32
LEARNING_RATE = 5e-3
LOSS_V = .5        # v loss coefficient
LOSS_ENTROPY = .01 # entropy coefficient

NUM_STATE = env.observation_space.shape[0]
NUM_ACTIONS = env.action_space.n
LOOPBACK_WINDOW = 4
NONE_STATE = np.zeros(NUM_STATE)

brain = Brain(LOOPBACK_WINDOW, 
              NUM_STATE, 
              NUM_ACTIONS, 
              LOSS_V, 
              LOSS_ENTROPY, 
              MIN_BATCH, 
              GAMMA_N, 
              LEARNING_RATE, 
              NONE_STATE) # brain is global in A3C

# Build the environments list according with thread count
envs=[]
for i in range(THREADS):
  trading_env = gym.make('TradeSim-v0')
  trading_env.env.configure(steps_per_episode=df_a3c_training.shape[0],      
                    instrument = 'EURUSD', 
                    equity=10000.0, 
                    leverage=1.0, 
                    commissions=0.5,
                    num_cdl_patterns=len(cdl_cols),
                    cb_pull_cdl_patterns=pull_last_cdl_patterns,
                    cb_pull_predictions=pull_predicted_data, 
                    cb_pull_ticks=pull_last_tick,
                    cb_market_info=None)
  e = Environment(trading_env, 
                  eps_start=EPS_START, 
                  eps_end=EPS_END, 
                  eps_steps=EPS_STEPS, 
                  num_state=NUM_STATE, 
                  num_actions=NUM_ACTIONS,
                  gamma = GAMMA,
                  gamma_n=GAMMA_N,
                  n_step_return=N_STEP_RETURN)
  envs.append(e)


# Build the optimizers list according with optimizers count
opts = [Optimizer() for i in range(OPTIMIZERS)]

# Starts each optimizer
for o in opts:
  o.start()

# Starts each environment
for e in envs:
  e.start()

# Waits RUN_TIME seconds
time.sleep(RUN_TIME)

# Stops each environment
for e in envs:
  e.stop()
# Kills each associated thread
for e in envs:
  e.join()

# Stops each optimizer
for o in opts:
  o.stop()
# Kills each optimizer thread
for o in opts:
  o.join()

print("Training finished")
````

Now let's provide a cell to execute a whole episode using Agent actions. Modifying variable `NUM_EPISODES` can execute multiple episodes in the same cell execution.

````python
# reset the environment
obs = env.reset()
done = False
NUM_EPISODES = 1
_episodes = 0
results={}

while _episodes < NUM_EPISODES:
  while not done:
    # Agent selects an action for current observation
    action = agent.selectAction(obs)
    # Executes the action
    state, reward, done, info = env.step(action)
    # Run Agent's learning process
    agent.learn(action, state, reward, done)
    
  # get episodic stats
  stats = env.env.statistics()
  # save episodic results
  results.append(stats)
  
  _episodes += 1
  
print('Episode loop Ends!')
````
