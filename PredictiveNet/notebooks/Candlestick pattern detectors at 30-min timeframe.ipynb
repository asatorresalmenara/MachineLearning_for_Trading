```python
# let's load historic M30 Bars in dataframe
df = utils.load_hist('../csv_data/EURUSD_M30.csv', sep=';', freq='30min', remove_weekends=True)
df.head()
```

```python
# Check first 1000 samples of CLOSE price
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.plot(df.CLOSE[0:1000])
# Left: with weekend data, right: removing weekend data
plt.subplot(1,2,2)
plt.plot(df.CLOSE.values[0:1000])
```

Let's add candlestick pattern detectors

```python
df = utils.add_candle_patterns(df)
df.head(2)
```

I will select only those columns starting by CDL

```python
cdl_cols = []
for c in df.columns:
  if c.startswith('CDL'):
    cdl_cols.append(c)
df = df[cdl_cols]
print('selected cols: ', cdl_cols)
```

```python
# Candlestick patterns visualization
rows = (df.shape[1] / 4) + 1
plt.figure(figsize = (12,50))
for c in df.columns:
  i = df.columns.get_loc(c)+1
  plt.subplot(rows, 4, i)
  plt.plot(df[c])
  plt.legend([c])
  
```

Now I filter those patterns less representative or with a small sample count, like: CDLUPSIDEGAP2CROWS

```python
df.drop(columns=['CDLUPSIDEGAP2CROWS'], inplace = True)
rows = (df.shape[1] / 4) + 1
plt.figure(figsize = (12,50))
for c in df.columns:
  i = df.columns.get_loc(c)+1
  plt.subplot(rows, 4, i)
  plt.plot(df[c])
  plt.legend([c])
  
```

All candlestick patterns are normalized to [-1, 1]. Let's list them and save in dataframe for later use.

```python
# Save file on disk
df.to_csv('../csv_data/EURUSD_M30_candlesticks.csv', sep=';')
print('File saved!')
```
